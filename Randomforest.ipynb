```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
F=pd.read_csv('C:\\Users\\82102\\Desktop\\Test\\income\\final1612.csv',encoding='euc-kr')
F
```

```python
F['class']=0
F1=F[['성별','연령','직업구분','평균소득','평균소비','평균처분가능소득','소득분위','소득대비소비','class']]
F1
```

```python
g=F1[F1['평균처분가능소득']<600].index
F1.drop(g,inplace=True)
F1.dropna(axis=0,inplace=True)
v=F1['평균처분가능소득']
```

```python
a=21000
b=11000
c=7000
d=4500
e=2700
f=600
```

```python
def get_score(v):
    if v >a:
        score = 'A'
    elif (v <=a) & (v >b):
        score = 'B'
    elif (v <=b) & (v >c):
        score = 'C'
    elif (v <=c) & (v >d):
        score = 'D'
    elif (v <=d) & (v >e):
        score = 'E'
    elif (v <=e) & (v >f):
        score = 'F'
    else:
        score = 'G'
    return score


F1["class"] = F1["평균처분가능소득"].apply(lambda v: get_score(v))
F1
```

```python
train, test = train_test_split(F1, test_size=0.2)
X_train = train[['소득대비소비','평균소득','평균소비','연령']]
y_train = train[['class']]
X_test = test[['소득대비소비','평균소득','평균소비','연령']]
y_test = test[['class']]

classifier = RandomForestClassifier(n_estimators = 100)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

print("정확도 : {}".format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.model_selection import GridSearchCV

grid = {
      'n_estimators' : [100,200],
      'max_depth' : [6,8,10,12],
      'min_samples_leaf' : [3,5,7,10],
      'min_samples_split' : [2,3,5,10]}

classifier_grid = GridSearchCV(classifier, param_grid = grid, scoring="accuracy", n_jobs=-1, verbose =1)
classifier_grid.fit(X_train, y_train)
print("최고 평균 정확도 : {}".format(classifier_grid.best_score_))
print("최고의 파라미터 :", classifier_grid.best_params_)
```
